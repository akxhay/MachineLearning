# Logistic Regression

## Part 1 - Data Preprocessing

### Importing the dataset

```python
import pandas as pd
dataset = pd.read_csv('data.csv')
```

```python
dataset.head(10)
```

### Getting the inputs and output

```python
X = dataset.iloc[:,1:-1].values
X
```

```python
Y = dataset.iloc[:,-1].values
Y
```

### Creating the Training Set and the Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size =0.2, random_state = 0)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_train
```

## Part 2 - Building and training the model

### Building the model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=0)
```

### Training the model

```python
model.fit(X_train, Y_train)
```

### Inference

Making the predictions of the data points in the test set

```python
Y_pred = model.predict(sc.transform(X_test))
```

```python
Y_pred
```

Making the prediction of a single data point with:

## Part 3: Evaluating the model

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(Y_test, Y_pred)
matrix
```

### Accuracy

```python
AR = (matrix[0][0] + matrix[1][1])/(matrix[0][0] + matrix[0][1]+ matrix[1][0] + matrix[1][1]) 
ER = (matrix[1][0] + matrix[0][1])/(matrix[0][0] + matrix[0][1]+ matrix[1][0] + matrix[1][1]) 
AR,ER
```

```python
from sklearn.metrics import accuracy_score
AR2 = accuracy_score(Y_test, Y_pred)
AR2
```
